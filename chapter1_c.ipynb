```python
%%writefile nested.c
#include <stdio.h>

int main()
{
    int marks;

    // Asking user to enter marks
    printf("Enter your marks: ");
    scanf("%d", &marks);

    // First condition (outer if)
    if (marks >= 0 && marks <= 100)
    {
        // Second condition (nested inside first if)
        if (marks >= 40)
        {
            printf("You are PASS\n");

            // Third condition (nested inside second if)
            if (marks >= 75)
            {
                printf("Grade: Distinction\n");
            }
            else
            {
                printf("Grade: Normal Pass\n");
            }
        }
        else
        {
            printf("You are FAIL\n");
        }
    }
    else
    {
        printf("Invalid marks entered\n");
    }

    return 0;
}
```
